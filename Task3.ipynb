2.Search query keyword analysis

```python
import string

query = "Buy mobile phone buy phone online"
query = query.lower()
query = query.translate(str.maketrans('', '', string.punctuation))
words = query.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

4. Email domain usage analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
#counting domains
domain_count = {}

for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)
#calculating percentages using percentage formula
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {int(percentage)}%")
```

5. Sales spike detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average = sum(sales) / len(sales)
threshold = average * 1.25

for day, value in enumerate(sales, start=1):
    if value >threshold:
        print(f"Day {day}: {value}")
```

6. Duplicate user id detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
#count occurrences
count_ids = {}
for uid in user_ids:
    count_ids[uid] = count_ids.get(uid, 0) + 1
#printing duplicates
for uid, count in count_ids.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
