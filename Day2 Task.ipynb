Unique words in a sentence

```python
sentence= input("Enter sentence:")
words= sentence.split()
unique_words= set(words)
print("unique words:",unique_words)
print("count of unique words:",len(unique_words))
```

2. Highest salary from Employee data

```python
employees = {}
n = int(input("Enter number of employees: "))
for i in range(n):
    name = input("Enter employee name: ")
    salary = float(input("Enter salary: "))
    employees[name] = salary
highest_employee = max(employees, key=employees.get)
print("\nEmployee Records:", employees)
print("Employee with highest salary:", highest_employee)
print("Highest salary:", employees[highest_employee])
```

3.Maximum and Minimum Values


```python
n = int(input("Enter number of elements: "))
numbers = []
for i in range(n):
    num = int(input("Enter number: "))
    numbers.append(num)
max = numbers[0]
min = numbers[0]
for num in numbers:
    if num > max:
        max = num
    if num < min:
        min = num
print("\nList:", numbers)
print("Max:", max)
print("Min:", min)
```

4. Count products above a price threshold

```python
n = int(input("Enter number of products: "))
prices = []
for i in range(n):
    price = int(input("Enter product price: "))
    prices.append(price)
count = 0
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000:", count)
```

5.Calculate Attendance percentage

```python
attendance =["P","P","A","P","P"]
present_ct=0
for day in attendance:
  if day== "P":
    present_ct+=1
total_days= len(attendance)
percentage= (present_ct/total_days)*100
print("attendance percentage=",percentage)
```

7.Count character frequency

```python
text = input("Enter a string: ")
char_count = {}
for ch in text:
    if ch in char_count:
        char_count[ch] += 1
    else:
        char_count[ch] = 1
print("Character counts:", char_count)
```

Calculate Average marks

```python
n = int(input("Enter number of subjects: "))
marks = []
for i in range(n):
    mark = float(input("Enter marks: "))
    marks.append(mark)
average = sum(marks) / len(marks)
print("Average Marks:", average)
```
