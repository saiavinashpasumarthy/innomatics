1.Mobile Recharge Validation

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Available plans:", valid_plans)
        return False
    else:
        print("Recharge Successful!")
        return True


# Retry system using while loop
while True:
    amt = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amt):
        break
    else:
        print("Please try again.\n")
```

2.Inventory reorder alert system

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(product, "→ Reorder Alert -need to be ordered immediately (Stock:", quantity, ")")
        else:
            print(product, "→ Stock OK (Stock:", quantity, ")")


# Sample Inventory data for real-time use
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 14,
    "Soap": 30,
    "Milk Powder": 8
}

# Function call
check_inventory(inventory)
```

3.Student result processing system

```python
def process_result(marks):
    total = 0
    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return average, "Pass"
    else:
        return average, "Fail"


# Input marks
student_marks = [60, 45, 70, 55, 40]
avg, result = process_result(student_marks)
print("Average Marks:", avg)
print("Result:", result)
```

4. Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    # peak hour extra charge
    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


# Retry system
while True:
    try:
        km = float(input("Enter distance travelled (km): "))
        peak_hour = input("Is it peak hour? (yes/no): ")
        if km <= 0:
            print("❌ Distance must be positive. Try again.\n")
            continue
        total_fare = calculate_fare(km, peak_hour)
        print("✅ Total Cab Fare: ₹", round(total_fare, 2))
        break
    except ValueError:
        print("❌ Invalid input. Please enter correct values.\n")
```

5. Employee Attendance Eligibility Checker


```python
def check_attendance(attendance):
    present_count = 0
    # count present days
    for day in attendance:
        if day.upper() == "P":
            present_count += 1
    percentage = (present_count / len(attendance)) * 100

    if percentage >= 75:
        return percentage, "Eligible"
    else:
        return percentage, "Not Eligible"
# Read attendance from user
attendance_list = input("Enter attendance (P/A separated by space): ").split()

percent, status = check_attendance(attendance_list)

print("Attendance Percentage:", percent)
print("Status:", status)
```

6. Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False

    # Rule 1: Minimum length
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    # validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Include digit and special character @#$)"
# User input
pwd = input("Enter password: ")
result = check_password(pwd)

print(result)
```
