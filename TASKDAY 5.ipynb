1.Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes):
    total_likes = 0

    for like in likes:      # loop to calculate total
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes_list = [200, 150, 300, 400]   # total = 1050
analyze_engagement(likes_list)
```

3.  Agriculture – Rainfall Adequacy Checker

```python
def analyze_rainfall(rainfall, required_level):
    total = 0

    # loop to calculate total rainfall
    for r in rainfall:
        total += r

    average = total / len(rainfall)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)
rainfall_data = [70, 75, 68, 80, 67]
analyze_rainfall(rainfall_data, 70)
```

4. Social Media – Duplicate Account Detection

```python
def detect_duplicates(usernames):
    unique_users = set()

    for user in usernames:
        if user in unique_users:
            print("Duplicate Accounts Found: Yes")
            return
        unique_users.add(user)

    print("Duplicate Accounts Found: No")


# Example usage
users = ["user1", "user2", "user3", "user1"]
detect_duplicates(users)
```

9. E-Learning – Student Attendance Eligibility System


```python
def check_exam_eligibility(attendance):
    total_days = len(attendance)
    present_days = 0

    # loop to count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / total_days) * 100

    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)
attendance_list = ["P", "P", "A", "P", "P"]  # 4/5 = 80%
check_exam_eligibility(attendance_list)
```

10.Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):
    bill = 0

    # billing slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status
units = int(input("Enter units consumed: "))

total_bill, usage_status = calculate_electricity_bill(units)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
