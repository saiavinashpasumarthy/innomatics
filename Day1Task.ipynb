```python
marks=[45,78,90,33,60]
passed = 0
fail = 0
for mark in marks:
  if mark >=50:
    passed += 1
  else:
    fail += 1
print("passed =", passed)
print("failed=", fail)
```

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_data=[]
for name in names:
  cleaned= name.strip().lower()
  cleaned_data.append(cleaned)
print("original names =",names)
print("cleaned names =",cleaned_data)
```
